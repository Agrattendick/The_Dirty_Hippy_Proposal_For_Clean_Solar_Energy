## Total Carbon Offset by State  
This plot shows the potential total carbon offset of each state if potential roofs had solar panels. It's a plot created off the base data found within Google's Project Sunroof dataset. This plot provides the baseline for the effects of landmass and population.

```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Import project sunroof data into pandas from CSV
sunroof_state = '../Resources/ProjectSunroofClean/project-sunroof-state-clean.csv'
sunroof_state_df = pd.read_csv(sunroof_state)

sunroof_state_df.head()
```

```python
# Bar Chart of Total Carbon Offset, and save as a png file.

#Plot the data
sunroof_state_df.plot.bar(x='state_name', y='carbon_offset_metric_tons', figsize = (25,10), color = 'green', legend = False,
                          width = 0.95)

#Titles
plt.title('Total Carbon Offset by State', fontsize=50)
plt.xlabel('State', fontsize = 50)
plt.ylabel('Carbon Offset in Metric Tons'
          '\n'
          '(Tens of Millions)', fontsize = 30)

#Xticks
plt.xticks(rotation=65, ha='right', fontsize=15)

#Horizontal Lines
increment = 10000000
line = list(range(1, 9))
a = 0
for i in line:
    plt.axhline(line[a]*increment, ls='--', color = 'black', lw = '1', zorder=0)
    a += 1

# Y-axis ticks
plt.tick_params(labelright=True)
plt.ylim(0,90000000)


#Save figure
plt.tight_layout()
#plt.savefig("../Images/total_carbon_offset_by_state.png")
```
